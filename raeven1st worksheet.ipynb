```python
#-----------------------------------
# 2.1.1 Use of Arithmetic operators
#-----------------------------------
a = 118
b = 32

c = a + b       # Use of + Addition
print (a, "+", b, "-->", c)

c = a - b       # Use of - Subtraction
print (a, "-", b, "-->", c)

c = a * b       # Use of * Multiplication
print (a, "*", b, "-->", c)

c = a / b       # Use of / Divison
print (a, "/", b, "-->", c)

c = a % b       # Use of % Modulus (remainder is displayed)
print (a, "%", b, "-->", c)


```

```python
#---------------------------------------------------
# 2.1.2 Use of Division (/) and floor division (//)
#---------------------------------------------------
a = 362
b = 64

c = a / b       # Use of /
print (a, " /", b, "-->", c)

c = a // b      # Use of //
print (a, "//", b, "-->", c)
```

```python
#----------------------------
# 2.1.3 Use of Power (**)
#----------------------------
a = 8
b = 3

c = a ** b       # Use of ** (Exponent)
print (a, "**", b, "-->", c)
```

```python
#-------------------------------
#2.2 Use of Bitwise Operators
#-------------------------------
# Use of & (and) operator
a = 0
b = 1
d = 1

c = a & b #Use of & (bitwise and)
print (a, "&", b, "-->", c)

c = d & b #Use of & (bitwise and)
print (d, "&", b, "-->", c)
```

```python
#-----------------------------
#2.2.2 Use of | (or) operator
#-----------------------------
a = 1
b = 1
d = 0

c = a | b      #Use of |
print (a, "|", b, "-->", c)

c = d | b      #Use of |
print (d, "|", b, "-->", c)
```

```python
#-------------------------------
#2.2.3: Use of ~ (not) operator
#-------------------------------
a = 0
b = 5

c = ~a #Use of ~
print ("~", a, "-->", c) #sign bit is used

c = ~b #Use of ~
print ("~", b, "-->", c) #sign bit is used
```

```python
#-----------------------
#2.2.4: Use of ^ (xor) operator
#-----------------------

a = 1
b = 1
d = 0

c = a ^ b #Use of ^
print (a, "^", b, "-->", c)

c = d ^ b #Use of ^
print (d, "^", b, "-->", c)
```

```python
#--------------------------------
#2.2.5: Use of >> (right shift) operators
#--------------------------------

a = 3
b = 2
d = 1

c = a >> b #Use of >>
print (a, ">>", b, "-->", c)

c = a >> d #Use of >>
print (a, ">>", d, "-->", c)
```
